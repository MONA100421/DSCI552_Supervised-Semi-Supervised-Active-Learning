````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, roc_curve,
    confusion_matrix
)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.exceptions import ConvergenceWarning
from sklearn.utils._testing import ignore_warnings

# === Step 1: Load and preprocess WDBC dataset ===
columns = ['ID', 'Diagnosis'] + [ 
    f'{feature}_{stat}' 
    for feature in ['radius', 'texture', 'perimeter', 'area', 'smoothness', 
                    'compactness', 'concavity', 'concave_points', 'symmetry', 
                    'fractal_dimension'] 
    for stat in ['mean', 'se', 'worst']
]
wdbc_df = pd.read_csv("../data/wdbc.data", header=None, names=columns)
X = wdbc_df.drop(['ID', 'Diagnosis'], axis=1).values
y = (wdbc_df['Diagnosis'] == 'M').astype(int).values  # Malignant = 1, Benign = 0

# === Step 2: Monte Carlo Simulation Parameters ===
M = 30
train_metrics = {'acc': [], 'prec': [], 'rec': [], 'f1': [], 'auc': []}
test_metrics = {'acc': [], 'prec': [], 'rec': [], 'f1': [], 'auc': []}

# === Step 3: Run 30 simulations ===
for run in range(M):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=run
    )
    
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    param_grid = {'C': np.logspace(-3, 3, 10)}
    cv = StratifiedKFold(n_splits=5)
    svm = LinearSVC(penalty='l1', dual=False, max_iter=10000)
    clf = GridSearchCV(svm, param_grid, cv=cv, scoring='f1', n_jobs=-1)
    clf.fit(X_train_scaled, y_train)

    best_model = clf.best_estimator_

    y_train_pred = best_model.predict(X_train_scaled)
    y_test_pred = best_model.predict(X_test_scaled)

    y_train_score = best_model.decision_function(X_train_scaled)
    y_test_score = best_model.decision_function(X_test_scaled)

    # Record metrics
    train_metrics['acc'].append(accuracy_score(y_train, y_train_pred))
    train_metrics['prec'].append(precision_score(y_train, y_train_pred))
    train_metrics['rec'].append(recall_score(y_train, y_train_pred))
    train_metrics['f1'].append(f1_score(y_train, y_train_pred))
    train_metrics['auc'].append(roc_auc_score(y_train, y_train_score))

    test_metrics['acc'].append(accuracy_score(y_test, y_test_pred))
    test_metrics['prec'].append(precision_score(y_test, y_test_pred))
    test_metrics['rec'].append(recall_score(y_test, y_test_pred))
    test_metrics['f1'].append(f1_score(y_test, y_test_pred))
    test_metrics['auc'].append(roc_auc_score(y_test, y_test_score))

# === Step 4: ROC Curve and Confusion Matrix from One Run ===
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=0
)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

clf.fit(X_train_scaled, y_train)
best_model = clf.best_estimator_
y_test_score = best_model.decision_function(X_test_scaled)
fpr, tpr, _ = roc_curve(y_test, y_test_score)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_test_score):.2f}")
plt.plot([0, 1], [0, 1], 'k--')
plt.title("ROC Curve (One Run)", fontsize=16, fontweight="bold")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.show()

# Confusion matrix
y_test_pred = best_model.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_test_pred)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Benign', 'Malignant'], yticklabels=['Benign', 'Malignant'])
plt.title("Confusion Matrix (Test Set)", fontsize=16, fontweight="bold")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# === Step 5: Print average metrics ===
print("### Average Metrics Over 30 Monte Carlo Runs ###")
for metric in train_metrics:
    print(f"{metric.upper()} - Train: {np.mean(train_metrics[metric]):.4f}, Test: {np.mean(test_metrics[metric]):.4f}")
````

#### (ii) Semi-Supervised Learning/ Self-training

##### (A)

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.semi_supervised import SelfTrainingClassifier

# Load the dataset
df = pd.read_csv("../data/data_banknote_authentication.txt", header=None)
df.columns = ['variance', 'skewness', 'curtosis', 'entropy', 'class']
X = df.drop('class', axis=1).values
y = df['class'].values

# Monte Carlo repetitions
M = 30
metrics = ['accuracy', 'precision', 'recall', 'f1-score', 'auc']
results = {m: [] for m in metrics}

# Repeat the experiment M times
for seed in range(M):
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=seed)

    # Normalize data
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Randomly select 50% positive and 50% negative as labeled
    pos_idx = np.where(y_train == 1)[0]
    neg_idx = np.where(y_train == 0)[0]
    np.random.seed(seed)
    labeled_pos = np.random.choice(pos_idx, len(pos_idx) // 2, replace=False)
    labeled_neg = np.random.choice(neg_idx, len(neg_idx) // 2, replace=False)
    labeled_idx = np.concatenate([labeled_pos, labeled_neg])
    unlabeled_idx = np.setdiff1d(np.arange(len(y_train)), labeled_idx)

    # Set unlabeled labels to -1
    y_train_semi = np.copy(y_train)
    y_train_semi[unlabeled_idx] = -1

    # Train L1-penalized SVM with 5-fold CV on labeled data
    svc = LinearSVC(penalty='l1', dual=False, max_iter=10000)
    grid = GridSearchCV(svc, param_grid={'C': [0.01, 0.1, 1, 10]}, cv=5)
    grid.fit(X_train_scaled[labeled_idx], y_train[labeled_idx])
    best_svc = grid.best_estimator_

    # Calibrate classifier to enable predict_proba
    calibrated_svc = CalibratedClassifierCV(best_svc, cv=3)
    calibrated_svc.fit(X_train_scaled[labeled_idx], y_train[labeled_idx])

    # Self-training with semi-supervised learning
    self_model = SelfTrainingClassifier(calibrated_svc, criterion='k_best', k_best=50)
    self_model.fit(X_train_scaled, y_train_semi)

    # Predictions and probability
    y_pred = self_model.predict(X_test_scaled)
    y_prob = self_model.predict_proba(X_test_scaled)[:, 1]

    # Evaluation metrics
    report = classification_report(y_test, y_pred, output_dict=True, zero_division=0)
    results['accuracy'].append(report['accuracy'])
    results['precision'].append(report['1']['precision'])
    results['recall'].append(report['1']['recall'])
    results['f1-score'].append(report['1']['f1-score'])
    results['auc'].append(roc_auc_score(y_test, y_prob))

# Final results table
final_result = pd.DataFrame({k: [np.mean(v)] for k, v in results.items()})
print("=== Self-training SVM (L1-Penalized) Average Results over 30 runs ===")
print(final_result.round(4))
````

I trained an L1-penalized SVM using only 50% of the positive and 50% of the negative examples in the training set as labeled data. I normalized all features using StandardScaler and selected the penalty parameter (C) via 5-fold cross-validation.

To handle the remaining unlabeled data, I used SelfTrainingClassifier with the best calibrated SVM. Over 30 Monte Carlo repetitions, I obtained the following average results:

| Metric     | Value   |
|------------|---------|
| Accuracy   | 0.9812  |
| Precision  | 0.9619  |
| Recall     | 0.9975  |
| F1-score   | 0.9793  |
| AUC        | 0.9996  |

These results confirm that the semi-supervised self-training approach is highly effective with only partially labeled data.

##### (B)

````python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix, roc_curve
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("../data/data_banknote_authentication.txt", header=None)
df.columns = ['variance', 'skewness', 'curtosis', 'entropy', 'class']
X = df.drop('class', axis=1).values
y = df['class'].values

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# Standardize
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Randomly select 50% positive & negative as labeled
np.random.seed(42)
pos_idx = np.where(y_train == 1)[0]
neg_idx = np.where(y_train == 0)[0]
labeled_pos = np.random.choice(pos_idx, len(pos_idx) // 2, replace=False)
labeled_neg = np.random.choice(neg_idx, len(neg_idx) // 2, replace=False)
labeled_idx = np.concatenate([labeled_pos, labeled_neg])
unlabeled_idx = np.setdiff1d(np.arange(len(y_train)), labeled_idx)

# Self-training loop
while len(unlabeled_idx) > 0:
    model = CalibratedClassifierCV(LinearSVC(C=1, penalty='l1', dual=False, max_iter=10000), cv=3)
    model.fit(X_train_scaled[labeled_idx], y_train[labeled_idx])
    
    probs = model.predict_proba(X_train_scaled[unlabeled_idx])
    uncertainty = np.abs(probs[:, 1] - 0.5)
    top_k = min(5, len(unlabeled_idx))
    uncertain_indices = np.argsort(uncertainty)[:top_k]
    chosen = unlabeled_idx[uncertain_indices]
    
    pseudo_labels = model.predict(X_train_scaled[chosen])
    y_train[chosen] = pseudo_labels
    labeled_idx = np.concatenate([labeled_idx, chosen])
    unlabeled_idx = np.setdiff1d(unlabeled_idx, chosen)

# Final model
final_model = CalibratedClassifierCV(LinearSVC(C=1, penalty='l1', dual=False, max_iter=10000), cv=3)
final_model.fit(X_train_scaled, y_train)

# Evaluation
y_train_pred = final_model.predict(X_train_scaled)
y_train_prob = final_model.predict_proba(X_train_scaled)[:, 1]
y_test_pred = final_model.predict(X_test_scaled)
y_test_prob = final_model.predict_proba(X_test_scaled)[:, 1]

train_report = classification_report(y_train, y_train_pred, output_dict=True)
test_report = classification_report(y_test, y_test_pred, output_dict=True)

results_df = pd.DataFrame({
    "Train Accuracy": [train_report["accuracy"]],
    "Train Precision": [train_report["1"]["precision"]],
    "Train Recall": [train_report["1"]["recall"]],
    "Train F1-Score": [train_report["1"]["f1-score"]],
    "Train AUC": [roc_auc_score(y_train, y_train_prob)],
    "Test Accuracy": [test_report["accuracy"]],
    "Test Precision": [test_report["1"]["precision"]],
    "Test Recall": [test_report["1"]["recall"]],
    "Test F1-Score": [test_report["1"]["f1-score"]],
    "Test AUC": [roc_auc_score(y_test, y_test_prob)]
})

print("=== Self-training SVM Final Evaluation ===")
print(results_df.to_string(index=False))

# Plot ROC curve and confusion matrices
train_cm = confusion_matrix(y_train, y_train_pred)
test_cm = confusion_matrix(y_test, y_test_pred)
fpr, tpr, _ = roc_curve(y_test, y_test_prob)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(fpr, tpr, label='ROC Curve', color='orange')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve (Test)", fontsize=16, fontweight="bold")
plt.legend()

plt.subplot(1, 3, 2)
sns.heatmap(train_cm, annot=True, fmt='d', cmap='Greens')
plt.title("Train Confusion Matrix", fontsize=16, fontweight="bold")
plt.xlabel("Predicted")
plt.ylabel("True")

plt.subplot(1, 3, 3)
sns.heatmap(test_cm, annot=True, fmt='d', cmap='Blues')
plt.title("Test Confusion Matrix", fontsize=16, fontweight="bold")
plt.xlabel("Predicted")
plt.ylabel("True")

plt.tight_layout()
plt.show()
````

To implement a **self-training semi-supervised learning** framework using `LinearSVC` as the base model. The model begins with a subset of labeled data (50% from each class), and iteratively labels the most uncertain examples from the unlabeled pool. The goal is to improve performance by leveraging both labeled and pseudo-labeled data.

#### Implementation Steps

1. **Data Loading**  
   - Source: `data_banknote_authentication.txt`  
   - Features: variance, skewness, curtosis, entropy  
   - Label: binary classification (0 = fake, 1 = authentic)

2. **Data Preprocessing**  
   - Stratified train-test split (80%/20%)  
   - Standardization using `StandardScaler`

3. **Self-training Strategy**  
   - Randomly select 50% from each class as initial labeled data  
   - Use `CalibratedClassifierCV(LinearSVC(...))` for classification with calibrated probabilities  
   - Iteratively:
     - Predict on unlabeled set
     - Choose `k=5` most uncertain samples (closest to 0.5 probability)
     - Assign pseudo-labels
     - Add to training pool
   - Repeat until all samples are labeled

4. **Final Training & Evaluation**  
   - Retrain SVM on the fully labeled dataset  
   - Evaluate on both training and test sets using:  
     - Accuracy  
     - Precision  
     - Recall  
     - F1-Score  
     - AUC

````python
from sklearn.cluster import KMeans
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay,
    roc_curve
)
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import numpy as np

# === Parameters ===
M = 10  # Number of runs
n_neighbors = 30

# === For tracking metrics ===
acc_list, prec_list, rec_list, f1_list, auc_list = [], [], [], [], []

for run in range(M):
    # Step 1: Fit KMeans
    kmeans = KMeans(n_clusters=2, init='k-means++', n_init=1, random_state=run)
    kmeans.fit(X_train_scaled)
    centers = kmeans.cluster_centers_
    labels = kmeans.labels_

    # Step 2: Assign labels via 30-NN majority vote
    cluster_label_map = {}
    nbrs = NearestNeighbors(n_neighbors=n_neighbors).fit(X_train_scaled)

    for cluster_id in range(2):
        center = centers[cluster_id].reshape(1, -1)
        _, indices = nbrs.kneighbors(center)
        neighbor_labels = y_train[indices[0]]
        majority_label = np.bincount(neighbor_labels).argmax()
        cluster_label_map[cluster_id] = majority_label

    # Step 3: Predict labels using cluster map
    predicted_labels = np.array([cluster_label_map[c] for c in labels])

    # Step 4: Evaluate
    acc_list.append(accuracy_score(y_train, predicted_labels))
    prec_list.append(precision_score(y_train, predicted_labels))
    rec_list.append(recall_score(y_train, predicted_labels))
    f1_list.append(f1_score(y_train, predicted_labels))
    auc_list.append(roc_auc_score(y_train, predicted_labels))

# === Print Average Results ===
print("=== Average Evaluation over M runs ===")
print(f"Accuracy : {np.mean(acc_list):.4f}")
print(f"Precision: {np.mean(prec_list):.4f}")
print(f"Recall   : {np.mean(rec_list):.4f}")
print(f"F1 Score : {np.mean(f1_list):.4f}")
print(f"AUC      : {np.mean(auc_list):.4f}")

# === Show metrics for one sample run (the last one) ===
cm = confusion_matrix(y_train, predicted_labels)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix", fontsize=16, fontweight="bold")
plt.grid(False)
plt.show()

fpr, tpr, _ = roc_curve(y_train, predicted_labels)
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f"AUC = {auc_list[-1]:.2f}", linewidth=2)
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve", fontsize=16, fontweight="bold")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
````

In this task, I use **KMeans clustering (k=2)** on the scaled training data and assign labels to each cluster by performing a **majority vote** among the **30 nearest neighbors** to each cluster center. These predicted labels are then compared to the ground-truth labels in the training data to compute the following evaluation metrics over **M = 10 runs**:

- **Accuracy**
- **Precision**
- **Recall**
- **F1-score**
- **ROC-AUC**

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_auc_score, roc_curve
)
from scipy.special import softmax
import seaborn as sns

# === Parameters ===
M = 10  # number of repeated runs
POSITIVE_CLASS = 1

# === Storage for metrics ===
accs, precs, recalls, f1s, aucs = [], [], [], [], []

# === Run M times ===
for seed in range(M):
    # Step 1: Train KMeans
    kmeans = KMeans(n_clusters=2, init='k-means++', n_init=10, random_state=seed)
    kmeans.fit(X_train_scaled)
    centers = kmeans.cluster_centers_

    # Step 2: Map cluster labels to true labels using training data
    train_preds = kmeans.predict(X_train_scaled)
    label_map = {}
    for cluster_id in np.unique(train_preds):
        true_labels = y_train[train_preds == cluster_id]
        majority_label = np.bincount(true_labels).argmax()
        label_map[cluster_id] = majority_label

    # Step 3: Predict test data by nearest center
    dists = np.linalg.norm(X_test_scaled[:, np.newaxis] - centers, axis=2)
    cluster_preds = np.argmin(dists, axis=1)
    label_preds = np.vectorize(label_map.get)(cluster_preds)

    # Step 4: Softmax probability for ROC/AUC
    probs = softmax(-dists, axis=1)[:, POSITIVE_CLASS]

    # Step 5: Evaluate
    accs.append(accuracy_score(y_test, label_preds))
    precs.append(precision_score(y_test, label_preds))
    recalls.append(recall_score(y_test, label_preds))
    f1s.append(f1_score(y_test, label_preds))
    aucs.append(roc_auc_score(y_test, probs))

    # Save final run results for confusion matrix & ROC
    if seed == M - 1:
        final_preds = label_preds
        final_probs = probs

# === Report average metrics ===
print(f"Average Performance over {M} runs (Positive class = {POSITIVE_CLASS}):")
print(f"Accuracy:  {np.mean(accs):.4f}")
print(f"Precision: {np.mean(precs):.4f}")
print(f"Recall:    {np.mean(recalls):.4f}")
print(f"F1-score:  {np.mean(f1s):.4f}")
print(f"AUC:       {np.mean(aucs):.4f}")

# === Plot confusion matrix (last run) ===
cm = confusion_matrix(y_test, final_preds)
plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix (Last Run)", fontsize=16, fontweight="bold")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# === Plot ROC curve (last run) ===
fpr, tpr, _ = roc_curve(y_test, final_probs)
plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, final_probs):.4f}")
plt.plot([0, 1], [0, 1], 'k--')
plt.title("ROC Curve (Last Run)", fontsize=16, fontweight="bold")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.show()
````

#### (iv) Spectral Clustering

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import SpectralClustering
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from collections import Counter
from sklearn.metrics.pairwise import rbf_kernel
import random

# === Step 1: Automatically select the best gamma that balances the two clusters ===
def find_best_gamma(X, y, gamma_list):
    p = np.sum(y == 1)
    n = np.sum(y == 0)
    total = len(y)
    target_ratio = p / total

    best_gamma = None
    min_diff = float('inf')

    for gamma in gamma_list:
        model = SpectralClustering(n_clusters=2, affinity='rbf', gamma=gamma, random_state=42)
        labels = model.fit_predict(X)
        cluster_counts = [np.sum(labels == i) for i in [0, 1]]
        ratio = min(cluster_counts) / sum(cluster_counts)
        diff = abs(ratio - target_ratio)

        if diff < min_diff:
            min_diff = diff
            best_gamma = gamma

    return best_gamma

# Define candidate gamma values
gamma_candidates = [0.1, 0.5, 1, 2, 5]
best_gamma = find_best_gamma(X_test, y_test, gamma_candidates)
print(f"Best gamma selected: {best_gamma}")

# === Step 2: Perform Spectral Clustering using the best gamma ===
spectral_model = SpectralClustering(
    n_clusters=2, 
    affinity='rbf', 
    gamma=best_gamma, 
    assign_labels='kmeans', 
    random_state=42
)
cluster_labels = spectral_model.fit_predict(X_test)

# === Step 3: Label each cluster using majority vote from 30 randomly sampled points ===
cluster_to_label = {}

for cluster_id in [0, 1]:
    cluster_indices = np.where(cluster_labels == cluster_id)[0]
    sample_indices = random.sample(list(cluster_indices), min(30, len(cluster_indices)))
    sample_labels = y_test[sample_indices]
    majority_label = Counter(sample_labels).most_common(1)[0][0]
    cluster_to_label[cluster_id] = majority_label

# === Step 4: Map cluster labels to predicted binary class labels ===
y_pred = np.vectorize(cluster_to_label.get)(cluster_labels)

# === Step 5: Evaluation ===
print("\n### Spectral Clustering Classification Report")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# === Step 6: Simulate soft probabilities using affinity scores for smoother ROC ===
affinity_matrix = rbf_kernel(X_test, gamma=best_gamma)
avg_affinity = affinity_matrix.mean(axis=1)

# Identify the cluster mapped to the positive class (label = 1)
positive_cluster = [cid for cid, label in cluster_to_label.items() if label == 1][0]

# Create soft score vector for ROC
soft_scores = (cluster_labels == positive_cluster).astype(int) * avg_affinity

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, soft_scores)
auc = roc_auc_score(y_test, soft_scores)

# Plot ROC curve
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}", lw=2)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Spectral Clustering (Soft Score)", fontsize=14, fontweight='bold')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
````

Use Spectral Clustering with an RBF kernel to classify test data into two balanced clusters, matching the original class distribution. Assign class labels via **majority vote** from 30 random points per cluster.

### Method Overview

* **Best Gamma Selection**: Chose gamma from `[0.1, 0.5, 1, 2, 5]` to match positive/negative ratio in ground truth.
* **Clustering**: Applied `SpectralClustering(...).fit_predict(X_test)`.
* **Label Assignment**: Each cluster labeled by majority class from 30 randomly sampled points.
* **Soft Scores**: Simulated probabilities using average RBF affinity per point.
* **ROC Curve**: Calculated AUC using soft scores.

#### (ii) Active

````python
import numpy as np
import pandas as pd
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Load the dataset (wdbc.data)
data = pd.read_csv("../data/wdbc.data", header=None)
X = data.iloc[:, 2:].values
y = data.iloc[:, 1].values
y = np.where(y == 'M', 1, 0)  # Convert 'M' to 1, 'B' to 0

# Standardize features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split into training (0.9) and testing (0.1)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, stratify=y, random_state=42)

# Convert training data into a pool
remaining_indices = list(range(len(X_train)))

# Initial pool has at least 2 classes
np.random.seed(42)
while True:
    initial_pool_idx = np.random.choice(remaining_indices, 10, replace=False)
    if len(np.unique(y_train[initial_pool_idx])) == 2:
        break

X_pool = X_train[initial_pool_idx]
y_pool = y_train[initial_pool_idx]

# Remove initial points from remaining pool
remaining_indices = list(set(remaining_indices) - set(initial_pool_idx))

test_errors = []
data_sizes = []

# Active Learning Loop
for i in range(90):
    # SVM with linear kernel and L1 penalty, use 5-fold CV to select best C
    param_grid = {'C': [0.01, 0.1, 1, 10]}
    svm = GridSearchCV(LinearSVC(penalty='l1', dual=False, max_iter=10000), param_grid, cv=5)
    svm.fit(X_pool, y_pool)

    # Predict and store test error
    y_pred = svm.predict(X_test)
    error = 1 - accuracy_score(y_test, y_pred)
    test_errors.append(error)
    data_sizes.append(len(X_pool))

    # Stop if no more data left
    if len(remaining_indices) == 0:
        break

    # Compute distance from hyperplane
    clf = svm.best_estimator_
    decision = clf.decision_function(X_train[remaining_indices])
    distances = np.abs(decision)

    # Find 10 closest points
    closest_idx = np.argsort(distances)[:10]
    selected_indices = [remaining_indices[i] for i in closest_idx]

    # Add to pool
    X_pool = np.vstack((X_pool, X_train[selected_indices]))
    y_pool = np.hstack((y_pool, y_train[selected_indices]))

    # Remove from remaining pool
    remaining_indices = list(set(remaining_indices) - set(selected_indices))

# Report
for size, err in zip(data_sizes, test_errors):
    print(f"Train size: {size}, Test error: {err:.4f}")
````

In this experiment, I implemented an active learning approach using a Linear SVM with L1 penalty on the Wisconsin Breast Cancer dataset. Starting with just 10 labeled samples containing both classes, I iteratively selected 10 new data points closest to the decision boundary using uncertainty sampling. At each step, I retrained the model with 5-fold cross-validation to find the best C and evaluated test error. As the labeled pool grew, the test error decreased significantly, from 14.04% at the beginning to 0.00% after incorporating all 512 training samples. This result demonstrates that active learning can achieve high accuracy with far fewer labeled examples by prioritizing the most informative ones.

### (c) Average the 50 test errors

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Load the dataset
data = pd.read_csv("../data/wdbc.data", header=None)  # Make sure the path is correct
X = data.iloc[:, 2:].values
y = np.where(data.iloc[:, 1] == 'M', 1, 0)  # Convert 'M' -> 1, 'B' -> 0

# Standardize the features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split into training and testing sets (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, stratify=y, random_state=42)

# Parameters for model tuning and simulation
param_grid = {'C': [0.01, 0.1, 1, 10]}
runs = 50          # Total simulation runs
max_iters = 90     # Number of iterations per run (adds 10 samples each)

# Lists to store error for each method
active_errors = []
passive_errors = []

for mode in ['active', 'passive']:
    method_errors = []

    for run in range(runs):
        np.random.seed(run)
        remaining_indices = list(range(len(X_train)))

        # Initialize pool with 10 samples (ensure both classes are present)
        while True:
            initial_pool = np.random.choice(remaining_indices, 10, replace=False)
            if len(np.unique(y_train[initial_pool])) == 2:
                break

        X_pool = X_train[initial_pool]
        y_pool = y_train[initial_pool]
        remaining_indices = list(set(remaining_indices) - set(initial_pool))

        run_test_errors = []

        for i in range(max_iters):
            if len(X_pool) < 2 or len(np.unique(y_pool)) < 2:
                break  # Ensure pool is valid for training

            # Train model with 5-fold CV to choose best C
            clf = GridSearchCV(LinearSVC(penalty='l1', dual=False, max_iter=20000), param_grid, cv=5)
            clf.fit(X_pool, y_pool)

            # Evaluate on test set
            y_pred = clf.predict(X_test)
            test_error = 1 - accuracy_score(y_test, y_pred)
            run_test_errors.append(test_error)

            if len(remaining_indices) == 0:
                break  # No more samples left

            # Select next 10 points
            k = min(10, len(remaining_indices))

            if mode == 'active':
                distances = np.abs(clf.best_estimator_.decision_function(X_train[remaining_indices]))
                selected_relative_indices = np.argsort(distances)[:k]
            else:  # passive
                selected_relative_indices = np.random.choice(len(remaining_indices), k, replace=False)

            selected_indices = [remaining_indices[i] for i in selected_relative_indices]

            # Add selected to pool and remove from remaining
            X_pool = np.vstack([X_pool, X_train[selected_indices]])
            y_pool = np.hstack([y_pool, y_train[selected_indices]])
            remaining_indices = list(set(remaining_indices) - set(selected_indices))

        # Record run results
        if mode == 'active':
            active_errors.append(run_test_errors)
        else:
            passive_errors.append(run_test_errors)

# Compute average test error for each iteration
avg_active = np.mean(active_errors, axis=0)
avg_passive = np.mean(passive_errors, axis=0)
data_sizes = list(range(10, 10 + 10 * len(avg_active), 10))

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(data_sizes, avg_active, marker='o', label='Active Learning')
plt.plot(data_sizes, avg_passive, marker='s', label='Passive Learning')
plt.xlabel("Number of Training Instances")
plt.ylabel("Average Test Error")
plt.title("Learning Curves: Active vs Passive Learning (50 runs)", fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
````
